```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o")
print(llm)
```

```python
result = llm.invoke("who is the president of the United States?")
print(result)
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an expert AI engineer. Solve my doubts."),
        ("user", "{input}")
    ]
)
prompt
```

```python
chain = prompt | llm
response = chain.invoke({"input": "how can i use AI foundry?"})
print(response)
```

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
chain = prompt | llm | parser
response = chain.invoke({"input": "how can i use AI foundry?"})
print(response)
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("FamilyGuide.pdf")
docs = loader.load()
print(docs)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=20,
    length_function=len,
    is_separator_regex=False,)

all_splits = text_splitter.split_documents(docs)

all_splits
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-large", dimensions = 1024)

embeddings
```

```python
from langchain_community.vectorstores import FAISS
import faiss

db = FAISS.from_documents(all_splits, embeddings)
db
```

```python
query = 'what are the visiting hours?'

docs = db.similarity_search(query)
docs[0].page_content
```

```python
retriever = db.as_retriever()
docs=retriever.invoke('what are the visiting hours?')
docs[0].page_content
```

```python
## save the faiss db
db.save_local("faiss_index")
```

```python
new_db = FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)
```

```python
new_retriever = new_db.as_retriever()
new_docs=new_retriever.invoke('what are the visiting hours?')
new_docs[0].page_content
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain

prompt = ChatPromptTemplate.from_template(
    """
Answer the following question/question's based on the provided context:
<context>
{context}
</context>

Question - {input}
"""
)

document_chain = create_stuff_documents_chain(llm, prompt)
document_chain
```

```python
from langchain_core.documents import Document

input = "what are the visiting hours and the requirements?"
contexts = retriever.invoke(input)

# Assuming contexts[0] is a Document object
context = contexts[0].page_content


# Pass the full context (which is a Document object, not just page_content)
document_chain.invoke({
    "input": input,
    "context": [Document(page_content=context)] # Pass the entire context object, not just page_content
})
```

```python
from langchain.chains import create_retrieval_chain

retrieval_chain = create_retrieval_chain(new_retriever, document_chain)
retrieval_chain.invoke({"input": input})['answer']
```

```python
from googletrans import Translator
translator = Translator()
translation = await translator.translate(retrieval_chain.invoke({"input": input})['answer'], dest='es')
translation.text
```
